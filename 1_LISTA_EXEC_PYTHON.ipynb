Escreva um programa que calcule o tempo estimado de uma viagem. O usuário deve informar a distância em km e a velocidade média esperada. O resultado deve ser exibido em horas e minutos

```python
d = float(input('Distância em km: '))
v = float(input('Velocidade média em km/h: '))
t = d / v
minutos = (round(t % 1, 2)) * 60
print('Tempo aproximado de %2.0f horas e %2.f minutos' %(t,
minutos))
```

```python
peso = float(input('Peso: '))

excesso = 0
multa = 0

if peso > 50:
    excesso = peso - 50
    multa = excesso * 4
else:
  print("Não há multas!!!")

print ('Excesso: %.2f' % excesso)
print ('Multa de R$ %.2f' % multa)
```

13. Maior e Menor de Três Números:

```python
a = int(input('a: '))
b = int(input('b: '))
c = int(input('c: '))
if a >= b and a >= c:
  print ('Maior: %d' %a)
elif b >= c:
  print ('Maior: %d' %b)
else:
  print ('Maior: %d' %c)
if a <= b and a <= c:
  print ('Menor: %d' %a)
elif b <= c:
  print ('Menor: %d' %b)
else:
  print ('Menor: %d' %c)
```

21. Vetor de Padrão Alternado (0 e 1):

Escreva um algoritmo que preencha um vetor de 20 posições com um padrão alternado: as posições de índice par devem receber o valor 1 e as posições de índice ímpar devem receber o valor 0. Imprima o vetor resultante na tela

```python
vetor = []
for i in range(20):
  if i % 2 == 0:
     vetor.append(1)
  else:
      vetor.append(0)
print("Vetor: ", vetor)
```
