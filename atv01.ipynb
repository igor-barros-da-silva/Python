3. Calculadora de Segundos.

Crie um programa que leia uma quantidade de dias, horas, minutos e segundos do usuário. Calcule e mostre o total de segundos correspondente a esses valores.

```python
d = int(input('Dias: '))
h = int(input('Horas: '))
m = int(input('Minutos: '))
s_input = input('Segundos: ')
s = int(s_input) if s_input.strip() else 0
total = d * 24 * 60 * 60 + h * 60 * 60 + m * 60 + s
print (total)
```

4. Cálculo de Aumento Salarial.

Faça um programa que solicite o valor do salário atual e a porcentagem de aumento. O programa deve calcular e imprimir o valor do aumento e o novo salário.

```python
s = float(input('Salário: '))
p = float(input('Porcentagem de aumento: '))
aumento = s * p / 100
novo = s + aumento
print('Aumento: R$ %.2f' %aumento)
print('Novo salário: R$ %.2f' %novo)
```

5. Cálculo de Desconto em Mercadoria.

Solicite o preço de uma mercadoria e o percentual de desconto. Calcule e exiba o valor do desconto e o preço final a pagar.

```python
m = float(input('Preço da mercadoria: '))
p = float(input('Porcentagem de desconto: '))
desconto = m * p / 100
novo = m - desconto
print('Desconto: R$ %.2f' %desconto)
print('Preço a pagar: R$ %.2f' %novo)
```

6. Tempo de Viagem.

Escreva um programa que calcule o tempo estimado de uma viagem. O usuário deve informar a distância em km e a velocidade média esperada. O resultado deve ser exibido em horas e minutos

```python
d = float(input('Distância em km: '))
v = float(input('Velocidade média em km/h: '))
t = d / v
minutos = (round(t % 1, 2)) * 60
print('Tempo aproximado de %2.0f horas e %2.f minutos' %(t,
minutos))
```

7. Aluguel de Carro

Desenvolva um algoritmo para calcular o preço do aluguel de um carro. O programa deve perguntar a quantidade de km percorridos e a quantidade de dias pelos quais o carro foi alugado. Calcule o preço total, sabendo que o carro custa R
 0,15 por km rodado.

```python
km = float(input('Km rodados: '))
dias = int(input('Dias: '))
preço = 60*dias + 0.15*km
print ('R$ %.2f' %preço)
```

8. Redução do Tempo de Vida de um Fumante.

Crie um programa para calcular a redução do tempo de vida de um fumante. Pergunte a quantidade de cigarros fumados por dia e quantos anos ele já fumou. Considere que um fumante perde 10 minutos de vida a cada cigarro. Calcule quantos dias de vida o fumante perderá e exiba o total.

```python
cigarros = int(input('Cigarros por dia: '))
anos = int(input('Anos fumados: '))
total_cigarros = anos * 365 * cigarros
dias_perdidos = total_cigarros / 144
print ('Você perdeu aproximadamente %.2f dias'
%dias_perdidos)
```
